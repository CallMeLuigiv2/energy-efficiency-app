```python

import pandas as pd
import numpy as np
import openpyxl

energy_df  =  pd.read_excel("./energy_DS.xlsx")

energy_df.head()
```

```python
rename_columns = {"X1":"Relative_Compactness","X2":"Surface_Area", "X3":"Wall_Area", "X4":"Roof_Area", "X5":"Overall_Height","X6":
                  "Orientation", "X7": "Glazing_Area","X8":"Glazing_Area_Distribution","Y1":"Heating_Load","Y2":"Cooling_Load"}

energy_df =  energy_df.rename(columns=rename_columns)
```

```python
energy_df.head(10)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10,8))
sns.heatmap(energy_df.corr(),annot=True, cmap="coolwarm")
plt.title("Checking Feature Correlations")
plt.show()
```

```python
sns.boxplot(x="Overall_Height" ,y= "Heating_Load", data=energy_df)
plt.show()
```

```python

sns.boxplot(x="Overall_Height" ,y= "Cooling_Load", data=energy_df)
plt.show()
```

```python
sns.histplot(energy_df["Heating_Load"], kde=True)
plt.title("Heating Load Distribution")
plt.show()
```

```python

sns.histplot(data=energy_df, x='Heating_Load', hue='Overall_Height', palette='viridis', multiple="stack")
plt.title("Heating Load split by Building Height")
plt.show()
```

```python
sns.histplot(energy_df["Cooling_Load"],kde=True)
plt.title("Cooling Load Distribution")
plt.show()
```

```python

sns.histplot(data=energy_df, x='Cooling_Load', hue='Overall_Height', palette='viridis', multiple="stack")
plt.title("Cooling Load split by Building Height")
plt.show()
```
